### using matplotlib

```python
x = [0,2,4,6,8]
y = [0,4,16,36,64]
fig, ax = plt.subplots()
ax.plot(x,y,marker = 'x',label = 'DATA POINTS')
ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()
plt.show()
print(x)
```
